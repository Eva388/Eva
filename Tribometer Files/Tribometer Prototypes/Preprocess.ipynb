```python
file = config.RAW_PATH + listdir(config.RAW_PATH)[1]
```

```python
def row_check(count):
    if count < 87: return 1
    elif count == 88: return 1
    elif count == 689: return 2
    elif (count > 689 and count < 699): return 1
    elif count == 700: return 1
    elif count >= 1301: return 3
    return 0
    
```

```python
with open(file,'r') as input_file:
    outs = []
    name = "test"
    name2 = "R_test"
    for num in range(1, config.NUM_SETS + 1):
        idx = str(num)
        outs.append(name + idx + ".csv")
        outs.append(name2 + idx + ".csv")
    
    print(outs)
    
    out_idx = 0
    count = 1
    for line in input_file:
        with open(config.CLEAN_PATH + outs[out_idx], 'w') as out_file:
            
                if row_check(count) == 1: 
                    count += 1
                    print(count,out_idx)
                    continue
                if row_check(count) == 2:
                    out_idx += 1
                    count += 1
                    print(count,out_idx)
                    continue
                if row_check(count) == 3:
                    out_idx += 1
                    count = 1
                    print(count,out_idx)
                    continue

                out_file.write(line)
                count += 1


    
```
